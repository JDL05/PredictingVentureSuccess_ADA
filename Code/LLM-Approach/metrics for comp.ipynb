```python
import pandas as pd 
from sklearn.metrics import classification_report, roc_auc_score

results_LLM = pd.read_csv('/Users/chris/Downloads/llm_predictions_results_f1optim.csv')
results_LLM 
```

```python
labels = results_LLM['actual_label']
predictions = results_LLM['predicted_label']
y_probs = results_LLM['success_probability']


report = classification_report(labels, predictions, target_names=["No-Success", "Success"], output_dict=True)

results = {
    "Model": "distilBERT LLM",
    "Accuracy": report["accuracy"],  # Overall accuracy
    "Precision": report["macro avg"]["precision"],  # Macro average precision
    "Recall": report["macro avg"]["recall"],  # Macro average recall
    "F1-Score": report["macro avg"]["f1-score"],  # Macro average F1-score
    "AUC": roc_auc_score(labels, y_probs)  # AUC score
}

results
```
