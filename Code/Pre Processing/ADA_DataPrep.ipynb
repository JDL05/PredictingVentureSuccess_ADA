### Remove unnecessary columns

```python
companies.drop(columns=['Organization Name URL', 'Operating Status', 'Stage', 'Headquarters Regions', 'Closed Date', 'Closed Date Precision', 'Company Type', 'Announced Date Precision', 'Price', 'Price Currency', 'Price (in USD)', 'Acquisition Terms', 'Money Raised at IPO', 'Money Raised at IPO Currency', 'Money Raised at IPO (in USD)', 'Valuation at IPO', 'Valuation at IPO Currency', 'Valuation at IPO (in USD)', 'Stock Symbol', 'Stock Symbol URL', 'Stock Exchange', 'Founded Date Precision', 'Exit Date Precision', 'Number of Events', 'IPO Date', 'Transaction Name', 'Postal Code', 'Transaction Name URL', 'Estimated Revenue Range', 'Headquarters Location', 'Last Funding Amount Currency', 'Last Equity Funding Amount Currency', 'Total Equity Funding Amount Currency', 'Total Funding Amount Currency'], inplace=True)
```

### Adjust NAs

```python
mean_founders = companies['Number of Founders'].mean()
companies['Number of Founders'] = companies['Number of Founders'].fillna(mean_founders)

companies = companies.dropna(subset=['Number of Employees'])
```

### Adjust Industry / Industry Categories 

```python
companies.loc[:, 'Industries'] = companies['Industries'].fillna('')
companies.loc[:, 'Industries'] = companies['Industries'].apply(lambda x: [item.strip() for item in x.split(',')])

companies.loc[:, 'Industry Groups'] = companies['Industry Groups'].fillna('')
companies.loc[:, 'Industry Groups'] = companies['Industry Groups'].apply(lambda x: [item.strip() for item in x.split(',')])
```

### Providing an ID

```python
companies.loc[:, 'ID'] = (companies.index + 1).astype(str).str.zfill(6)

columns_order = ['ID'] + [col for col in companies.columns if col != 'ID']
companies = companies[columns_order]
```

## Create new People Table

```python
temp_companies = companies.copy()

temp_companies['Founders'] = temp_companies['Founders'].fillna('').astype(str)
temp_companies['Founders'] = temp_companies['Founders'].apply(lambda x: [item.strip() for item in x.split(',') if item.strip()])

founders_table = temp_companies.explode('Founders', ignore_index=True)

founders = founders_table[['ID', 'Founders', 'Organization Name']].rename(
    columns={'ID': 'Company ID', 'Founders': 'Founder Name'}
)

founders['Founder ID'] = founders.index + 1
founders['Founder ID'] = founders['Founder ID'].astype(str).str.zfill(6)

founders = founders[['Founder ID', 'Company ID', 'Founder Name', 'Organization Name']]

founders = founders.dropna(subset=['Founder Name'])
companies = companies.dropna(subset=['Number of Employees'])
```

## Enhance founders with people table
Via the Crunchbase API we received access to a csv file with additional people data. The file is already imported with the name 'people'. Now, we enhance the newly created data set founders with the additional data.

```python
founders[['first_name', 'last_name']] = founders['Founder Name'].str.split(' ', n=1, expand=True)

founders['linkedin_url'] = None
duplicates = [] 

for idx, row in founders.iterrows():
    matches = people[
        (people['first_name'] == row['first_name']) &
        (people['last_name'] == row['last_name'])
        ]
    if len(matches) == 0:
        continue
    elif len(matches) == 1:
        match = matches.iloc[0]

        founders.at[idx, 'linkedin_url'] = match['linkedin_url']
        
        for col in ['logo_url', 'facebook_url', 'twitter_url', 'city', 'region', 'country_code',
                    'featured_job_title', 'featured_job_organization_name']:
            founders.at[idx, col] = match[col]
    else:

        duplicates.append({
            'founder_id': row['Founder ID'],  
            'first_name': row['first_name'],
            'last_name': row['last_name'],
            'matches': matches.to_dict('records')  
        })

print("\nResult with duplicates:")
print(founders)

print("\nList of Duplicates:")
for duplicate in duplicates:
    print(f"Founder ID: {duplicate['founder_id']}, Name: {duplicate['first_name']} {duplicate['last_name']}")
    print("Matches:")
    for match in duplicate['matches']:
        print(f"  - LinkedIn: {match['linkedin_url']}")
```

## Investors
We use the investors data set to determine the quality of the investor group as larger funds may have better access to startups and provide a positive signalling for the startup. Example: If Sequoia is investing in a startup, many other investors and industry experts will evaluate the startup for a potential future round and give legitimacy for other investors.

In our case, we have a list of investors and their activity from multiple countries: Germany (GER), Austria (AUS), Switzerland (SWI), United Kingdom (UK), United States of America (USA). We calculate a investor score for each startup based on a self-defined heuristic approach.

```python
investors_AUS['Country'] = 'AUS'
investors_GER['Country'] = 'GER'
investors_SWI['Country'] = 'SWI'
investors_UK['Country'] = 'UK'
investors_USA['Country'] = 'USA'

all_investors = pd.concat([investors_AUS, investors_GER, investors_SWI, investors_UK, investors_USA], ignore_index=True)
```
